```python
# Split the data into train and test
# There are 600 tweets and we must build a model using 500, so we keep 100 for testing
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

train_tweets, test_tweets = train_test_split(train_test_tweets, test_size=1/6)
train_tweets.head(20)
```

```python
train_x=train_tweets['Tweet'].tolist()
train_y=train_tweets['Sentiment'].tolist()
train_x
```

```python
# Remove punctuation, convert to lowercase 

def clean(text):
    # Remove punctuation
    remove_punc=[char for char in text if char not in string.punctuation]
    remove_punc=''.join(remove_punc)
    cleaned=[word.lower() for word in remove_punc.split() if len(word)>1]
    return cleaned

clean('This is a test sentence!!!')
```

```python
train_x_reformat=[clean(row) for row in train_x]
train_x_reformat
```

```python
# Compile list of all words 
train_x_words=[]
for e in train_x_reformat:
    for f in e:
        train_x_words.append(f)
        
train_x_words
```

```python
# Apply FreqDist
import nltk
fd=nltk.FreqDist(train_x_words)
train_x_unique_words=[e for e in fd.keys()]
train_x_unique_words
```

```python
# Create feature extraction to detect presence of words

def feature_extraction(text):
    features_extracted={}
    for e in train_x_unique_words:
        features_extracted[e] = e in text
    return features_extracted

feature_extraction("cannabis is cool i support it")
```

```python
model_input=[(feature_extraction(train_x_reformat[i]), train_y[i]) for i in range(len(train_y)-1)]
model_input
```

```python
NBclassifier=nltk.NaiveBayesClassifier.train(model_input)
```

```python
NBclassifier.classify(feature_extraction("i am not happy about this"))
```

```python
# Reformat the test tweets 
test_x=test_tweets['Tweet'].tolist()
test_y=test_tweets['Sentiment'].tolist()
test_x
```

```python
test_x_reformat=[clean(row) for row in test_x]
test_x_reformat
```

```python
# Accuracy 
count=0
correct=0
for i in range(len(test_y)-1):
    predict=NBclassifier.classify(feature_extraction(test_x_reformat[i]))
    actual=test_y[i]
    if predict==actual:
        correct=correct+1
    count=count+1
correct/count
    
```
