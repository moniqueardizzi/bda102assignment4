```python
# Checking dimensions
train_test_tweets.shape
```

```python
# Split the data into train and test
# There are 600 tweets and we must build a model using 500, so we keep 100 for testing
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

train_tweets, test_tweets = train_test_split(train_test_tweets, test_size=1/6)
# Check on train 
train_tweets.head(20)
```

```python
# Checking that train has 500 tweets
train_tweets.shape
```

```python
# Checking that test has 100 tweets
test_tweets.shape
```

```python
# Convert the tokenized and cleaned training tweets and labelled sentiments to lists 
train_x=train_tweets['tokenized_cleaned'].tolist()
train_y=train_tweets['Sentiment'].tolist()
train_x[0:9]
train_y[0:9]
```

```python
# Create list of tuples containing list of tweet words and the labelled sentiment 
reformat_train=[]
for i in range(0,len(train_x)):
    wordlist=train_x[i].split()
    sentiment=train_y[i]
    reformat_train.append((wordlist, sentiment))

reformat_train[0:5]
```

```python
# Compile list of all words 
train_words=[]
for (x,y) in reformat_train:
    train_words.extend(x)
        
train_words[0:20]
```

```python
# Apply FreqDist to get list of unique words to build a feature extraction function 
import nltk
fd=nltk.FreqDist(train_words)
train_unique_words=[e for e in fd.keys()]
train_unique_words[0:20]
```

```python
# Create feature extraction to detect presence of words

def feature_extraction(text):
    features_extracted={}
    for e in train_unique_words:
        features_extracted[e] = e in text
    return features_extracted

# Test case 
feature_extraction("i use marijuana")
```

```python
# Create the model input - the feature extraction result and the labelled sentiment 
model_input=[(feature_extraction(words), sentiment) for (words, sentiment) in reformat_train]
model_input[0]
```

```python
# Create the Naive Bayes classifer 
SentimentNBclassifier=nltk.NaiveBayesClassifier.train(model_input)
```

```python
# Check that the labels assigned by the classifier cover positive, negative, neutral 
SentimentNBclassifier.labels()
```

```python
# We can also check the most predictive features based on the training data
SentimentNBclassifier.show_most_informative_features()
```

```python
# Reformat the test tweets so that we can check the accuracy of our model 

test_x=test_tweets['tokenized_cleaned'].tolist()
test_y=test_tweets['Sentiment'].tolist()
test_x[0:9]
test_y[0:9]
```

```python
reformat_test=[]
for i in range(0,len(test_x)):
    wordlist=test_x[i].split()
    sentiment=test_y[i]
    reformat_test.append((wordlist, sentiment))

reformat_test[0:5]
```

```python
# Create test input - features extracted and labelled sentiment 
test_input=[(feature_extraction(words), sentiment) for (words, sentiment) in reformat_test]
```

```python
# Our model has 76% accuracy (76 out of 100 cases were predicted correctly)
print(nltk.classify.accuracy(SentimentNBclassifier, test_input))
```
